# Spliting train_test data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
# Using onehotencoder for preprocessing
df = pd.get_dummies(df, columns = ['Grade'])
df
```

```python
x = df[['Grade_FAQ', 'Grade_Large', 'Grade_Medium', 'Grade_Small', 'Min Price', 'Max Price']]
y = df['Modal Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
x_train
```

```python
x_test
```

# Pridection of data

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Linear_Regrassion_Model
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
linear_pred = linear_model.predict(x_test)
```

```python
linear_pred
```

```python
# Checking Score
linear_score = mean_squared_error(y_test, linear_pred)
linear_score
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
linear_mape = mean_absolute_percentage_error(y_test, linear_pred, sample_weight = None, multioutput = 'uniform_average')
linear_mape
```

```python
# DecisionTreeRegressor_model
decision_model = DecisionTreeRegressor()
decision_model.fit(x_train, y_train)
```

```python
decision_y = decision_model.predict(x_test)
decision_y
```

```python
# MSE
decision_mse = mean_squared_error(y_test, decision_y)
print('Score:', decision_mse)
```

```python
# MAPE
decision_mape = mean_absolute_percentage_error(y_test, decision_y, sample_weight = None, multioutput = 'uniform_average')
decision_mape
```

# Conclusion
## Over Linear_model has good accuracy than DecisionTree

```python
Score = {'Linear': [linear_mape], 'DecisionTree': [decision_mape]}
Score = pd.DataFrame(Score)
Score
```
